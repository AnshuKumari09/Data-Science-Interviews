### Working with Dates

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1. Extract year

```python
date['date_year']=date['date'].dt.year
```

```python
date['date_year']
```

### 2. Extract Month

```python
date['date_month']=date['date'].dt.month
```

```python
date['date_month']
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week
date['date_day_name'] = date['date'].dt.day_name()

date.head()
```

```python
date.drop(columns=['product_id','city_id','orders'])
```

```python
date['date_is_weekend']=np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date['date_is_weekend']
```

### Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date
```

```python
date['quarter']=date['date'].dt.quarter
```

```python
# ensure datetime type
time['date'] = pd.to_datetime(time['date'], errors='coerce')

# now extract only time (hh:mm:ss)
time['time'] = time['date'].dt.time

time.head()
```
