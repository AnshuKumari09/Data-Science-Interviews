```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

drop='first'
Jab hum categorical variable ko dummy variables me convert karte hai (jaise fuel = {Petrol, Diesel, CNG}), to by default 3 columns ban jaate hai.
Lekin actual me sirf 2 column hi kaafi hote hai kyunki teesra column automatically infer ho jata hai.
ðŸ‘‰ Isko dummy variable trap bolte hai (multicollinearity problem).
Fuel: Petrol, Diesel, CNG

Without drop='first':
Petrol â†’ [1,0,0]
Diesel â†’ [0,1,0]
CNG    â†’ [0,0,1]

With drop='first':
Petrol â†’ [0,0]
Diesel â†’ [1,0]
CNG  
sparse=False
Kuch encoders sparse matrix (compressed memory structure) return karte hai to bachane ke liye space. âœ… sparse=False â†’ output normal dense array me aayega.dtype=np.int32 â†’ memory bachane ke liye 32-bit integers me store karega.  â†’ [0,1]


```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
X_train_new=ohe.fit_transform(X_train[['fuel','owner']])  #fit() â†’ categories ko seekh leta hai (unique values nikalta hai, jaise "Petrol", "Diesel", "CNG", "First Owner", "Second Owner" etc.).
#transform() â†’ unhe numerical format me convert karta hai (0 aur 1 ke form me â†’ One Hot Encoding).
```

### 4. OneHotEncoding with Top Categories

```python
df['brand'].nunique()
```

```python
threshold=100
counts=df['brand'].value_counts()
```

```python
counts
```

threshold=100 set kiya matlab: Jinke cars 100 se kam hain â†’ unhe rare brand maana jayega.
repl me un rare brands ke naam store ho gaye. BMW (50 cars),Skoda (40 cars),Jaguar (10 cars)
sab aa jayenge repl me.

```python
repl=counts[counts<=threshold].index
```

```python
df['brand'].replace(repl,'Uncommon') #Rare brands ko replace kar diya ek hi label se: "uncommon".
```
