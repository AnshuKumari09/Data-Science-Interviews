2️⃣ df.iloc[:,0:4] क्या कर रहा है?
iloc[:,0:4] → dataframe से 0th से 3rd column तक select कर रहा है (यानी 4 columns)।
ये होते हैं features (X) = input variables जिनसे prediction करनी है।

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
X_train.head()
```

y_train

```python
from sklearn.preprocessing import OneHotEncoder
```

drop='first'
Jab hum categorical variable ko dummy variables me convert karte hai (jaise fuel = {Petrol, Diesel, CNG}), to by default 3 columns ban jaate hai.
Lekin actual me sirf 2 column hi kaafi hote hai kyunki teesra column automatically infer ho jata hai.
👉 Isko dummy variable trap bolte hai (multicollinearity problem).
Fuel: Petrol, Diesel, CNG

Without drop='first':
Petrol → [1,0,0]
Diesel → [0,1,0]
CNG    → [0,0,1]

With drop='first':
Petrol → [0,0]
Diesel → [1,0]
CNG  
sparse=False
Kuch encoders sparse matrix (compressed memory structure) return karte hai to bachane ke liye space. ✅ sparse=False → output normal dense array me aayega.dtype=np.int32 → memory bachane ke liye 32-bit integers me store karega.  → [0,1]


```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
X_train_new=ohe.fit_transform(X_train[['fuel','owner']])  #fit() → categories ko seekh leta hai (unique values nikalta hai, jaise "Petrol", "Diesel", "CNG", "First Owner", "Second Owner" etc.).
#transform() → unhe numerical format me convert karta hai (0 aur 1 ke form me → One Hot Encoding).
```

### 4. OneHotEncoding with Top Categories

```python
df['brand'].nunique()
```

```python
threshold=100
counts=df['brand'].value_counts()
```

```python
counts
```

threshold=100 set kiya matlab: Jinke cars 100 se kam hain → unhe rare brand maana jayega.
repl me un rare brands ke naam store ho gaye. BMW (50 cars),Skoda (40 cars),Jaguar (10 cars)
sab aa jayenge repl me.

```python
repl=counts[counts<=threshold].index
```

```python
df['brand'].replace(repl,'Uncommon') #Rare brands ko replace kar diya ek hi label se: "uncommon".
```

pd.get_dummies(...) kya karta hai?Ye one-hot encoding hai. Har unique category ek naya column ban jata (BMW, Hyundai, Maruti…).
Agar row ka brand BMW hai → to BMW=1, baaki sab 0. Agar row ka brand Toyota hai → to Toyota=1, baaki sab 0.

| brand   | BMW | Hyundai | Maruti | Toyota | Uncommon |

| Kia     | 0   | 0       | 0      | 0      | 1        |
