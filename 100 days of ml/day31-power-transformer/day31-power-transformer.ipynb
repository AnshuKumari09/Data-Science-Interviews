```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt=PowerTransformer(method='box-cox')
```

PowerTransformer ek sklearn ka transformer hai jo power transformations apply karta hai (Box-Cox ya Yeo-Johnson).Yaha method='box-cox' use kiya gaya hai.
Box-Cox transform ke liye condition hai ki input values > 0 honi chahiye. Isliye aage +0.000001 kiya gaya hai (taaki 0 ya negative na ho jaye).

```python
X_train_transformed=pt.fit_transform(X_train+0.000001)
```

```python
X_test_tranformed=pt.transform(X_test+0.000001)
```

fit_transform training data par lagaya gaya, matlab: pehle ye λ (lambda) parameter estimate karta hai jo Box-Cox ke liye best hota hai.
phir transformation apply karta hai. transform test set par wahi learned transformation apply karta hai (same λ use hota hai, naya λ calculate nahi hota).

### Ye pura code check karta hai ki X_train ke columns ko kaise transform karein (Box-Cox ke saath) taaki wo normal distribution ke close ho jayein. Har column ka apna ek best λ hota hai, jo pt.lambdas_ me milta hai. Fir wahi transform X_test par bhi lagaya gaya hai.

pt.lambdas_ har feature ke liye ek λ (lambda) deta hai.
Box-Cox formula hai:

## 𝑦(𝜆)={(𝑥^𝜆−1)/𝜆,  𝜆≠0
###        { ln(𝑥),      𝜆 =0
Ye λ batata hai ki data ko normal distribution ke kareeb lane ke liye kaunsa power transformation best hai.
λ = 1 → data waise ka waise reh gaya (kuch change nahi)
λ = 0 → log transform ke barabar
0 < λ < 1 → square root jaise transformation
λ > 1 → variance reduce karne ke liye stronger transform

```python
pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
# Apply Yeo-Johnson transform

pt1 = PowerTransformer()

X_train_transformed2 = pt1.fit_transform(X_train)
X_test_transformed2 = pt1.transform(X_test)

lr = LinearRegression()
lr.fit(X_train_transformed2,y_train)

y_pred3 = lr.predict(X_test_transformed2)

print(r2_score(y_test,y_pred3))

pd.DataFrame({'cols':X_train.columns,'Yeo_Johnson_lambdas':pt1.lambdas_})
```

```python
# applying cross val score

pt = PowerTransformer()
X_transformed2 = pt.fit_transform(X)

lr = LinearRegression()
np.mean(cross_val_score(lr,X_transformed2,y,scoring='r2'))
```

```python
X_train_transformed2 = pd.DataFrame(X_train_transformed2,columns=X_train.columns)
```

```python
# Before and after comparision for Yeo-Johnson

for col in X_train_transformed2.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(X_train[col])
    plt.title(col)

    plt.subplot(122)
    sns.distplot(X_train_transformed2[col])
    plt.title(col)

    plt.show()
```

```python
# Side by side Lambdas
pd.DataFrame({'cols':X_train.columns,'box_cox_lambdas':pt.lambdas_,'Yeo_Johnson_lambdas':pt1.lambdas_})
```
