```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# üîπ Scenario:

# Task: MNIST digit classification (0‚Äì9)

# Input: 28√ó28 images ‚Üí flatten karke 784 features

# Network:

# Input layer ‚Üí 784 neurons

# Hidden layer 1 ‚Üí 128 neurons

# Hidden layer 2 ‚Üí 32 neurons

# Output layer ‚Üí 10 neurons (classes 0‚Äì9)
```

```python
import numpy as np
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
model=Sequential()
```

```python
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python

history = model.fit(X_train,y_train,epochs=25,validation_split=0.2)
```

```python
y_prob = model.predict(X_test)
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'], label='Training Loss')  #üëâ Ye batata hai model kitna galat predict kar raha hai training data par.Loss = "error" (jitna chhota, utna accha).
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.plot(history.history['accuracy'], label='Training Accuracy') #üëâ Ye batata hai model training data par kitna sahi predict kar raha hai.
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')

plt.title('Model Training Progress')
plt.xlabel('Epochs')
plt.ylabel('Value')
plt.legend()   # <-- Yeh show karega kaunsi line kis cheez ki hai
plt.show()
```
